```python
# Convert PatientId to string
df['PatientId'] = df['PatientId'].astype(str)

# Convert date columns to datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
df = df[df['Age'] >= 0]         # Remove negative ages
df = df[df['Age'] <= 115]       # Remove unrealistic old ages
```

```python
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'No-show': 'No_Show'
}, inplace=True)
```

```python
df['No_Show'] = df['No_Show'].str.strip().str.title()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# --- 1️⃣ Detect Outliers in Age ---
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Box Plot - Age')
plt.show()

# Find Age outliers
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

age_outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
print("Age Outliers:\n", age_outliers)
```

```python
# --- 3️⃣ Remove Outliers (Example for Age) ---
df_cleaned = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
```

```python
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")
```

```python
df_cleaned
```

```python
import pandas as pd
# 1️⃣ Drop irrelevant columns
df = df.drop(columns=['Scholarship', 'Neighbourhood'], errors='ignore')

# 2️⃣ Rename columns for clarity
df = df.rename(columns={
    'Hipertension': 'Hypertension',
    'Diabetes': 'Has_Diabetes',
    'SMS_received': 'SMS_Received',
    'No-show': 'No_Show'
})

# 3️⃣ Create a new column: Age Group
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[0, 12, 19, 35, 50, 65, 100],
                         labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior'])

# 4️⃣ Convert 'No_Show' to Boolean (0 = Showed, 1 = No-show)
df['No_Show_Flag'] = df['No_Show'].map({'No': 0, 'Yes': 1})

# 5️⃣ Filter: Patients older than 60 who didn’t show up
senior_no_show = df[(df['Age'] > 60) & (df['No_Show_Flag'] == 1)]

# 6️⃣ Sort: By Age (descending)
df_sorted = df.sort_values(by='Age', ascending=False)

# 7️⃣ Group & Aggregate: No-show rate by Age Group
no_show_rate = df.groupby('Age_Group')['No_Show_Flag'].mean().reset_index()

# Display results
print("Senior Patients Who Missed Appointments:\n", senior_no_show.head())
print("\nNo-Show Rate by Age Group:\n", no_show_rate)

# Save manipulated dataset
df.to_csv("manipulated_dataset.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load manipulated dataset
df = pd.read_csv("manipulated_dataset.csv")
```

```python
df
```

```python
# 1️⃣ Bar Plot - No-show rate by Age Group
no_show_rate = df.groupby('Age_Group')['No_Show_Flag'].mean().reset_index()
plt.figure(figsize=(8,5))
plt.bar(no_show_rate['Age_Group'], no_show_rate['No_Show_Flag'], color='skyblue')
plt.title("No-show Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("No-show Rate")
plt.show()
```

**No-show Rate by Age Group –** Shows the proportion of missed appointments for different age groups, highlighting that teens have the highest no-show rate while seniors have the lowest.

```python
# 3️⃣ Pie Chart - Show vs No-show
show_counts = df['No_Show'].value_counts()
plt.pie(show_counts, labels=show_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral'])
plt.title("Show vs No-show Ratio")
plt.show()
```

**Show vs No-show Ratio:** About 79.8% of patients showed up for their appointments, while 20.2% missed them.

```python
# ===================== SEABORN VISUALIZATIONS =====================

# 5️⃣ Scatter Plot - Age vs No-show
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='No_Show_Flag', data=df, alpha=0.5)
plt.title("Age vs No-show")
plt.show()
```

**Age vs No-show:** No-show behavior is spread across all ages, with a few outliers near age 0 and over 100.

```python
# 6️⃣ Heatmap - Correlation
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for Gender vs Hypertension
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", hue="Hypertension", data=df, palette="Set2")
plt.title("Gender vs Hypertension")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Hypertension", labels=["No", "Yes"])
plt.show()
```

**Gender vs Hypertension:** Females have higher counts in both hypertensive and non-hypertensive groups compared to males.

```python
# Countplot for Gender vs Has_Diabetes
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", hue="Has_Diabetes", data=df, palette="Set1")
plt.title("Gender vs Has_Diabetes")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Has_Diabetes", labels=["No", "Yes"])
plt.show()
```

**Gender vs Has_Diabetes:** Females have higher counts in both Has_Diabetese and non-Has_Diabetes groups compared to males.

```python
# Box plot: Age vs Hypertension
plt.figure(figsize=(8,5))
sns.boxplot(x='Hypertension', y='Age', data=df, palette='Set2')
plt.title('Age vs Hypertension')
plt.xlabel('Hypertension (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

This box plot compares the age distribution for two groups: those with hypertension (1) and those without (0). The group with hypertension has a higher median age and a wider range of ages, indicating that older individuals are more likely to have the condition.

```python
# Box plot: Age vs Has_Diabetes
plt.figure(figsize=(8,5))
sns.boxplot(x='Has_Diabetes', y='Age', data=df, palette='Set3')
plt.title('Age vs Has_Diabetes')
plt.xlabel('Has_Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

Similar to the hypertension plot, this box plot illustrates that the median age is significantly higher for individuals with diabetes (1) compared to those without (0). This suggests that diabetes is more prevalent in older populations.



```python
# Create Age Groups
bins = [0, 18, 30, 45, 60, 75, 100]
labels = ['0-17', '18-29', '30-44', '45-59', '60-74', '75+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
# Bar plot: Hypertension by Age Group
plt.figure(figsize=(8,5))
sns.barplot(x='AgeGroup', y='Hypertension', data=df, ci=None, palette='Blues')
plt.title('Hypertension Rate by Age Group')
plt.ylabel('Average (Proportion)')
plt.show()
```

This bar chart shows a direct relationship between age and the average proportion of people with hypertension. As age increases, the rate of hypertension also increases significantly, with the highest proportion seen in the 75+ age group.



```python
# Bar plot: Has_Diabetes by Age Group
plt.figure(figsize=(8,5))
sns.barplot(x='AgeGroup', y='Has_Diabetes', data=df, ci=None, palette='Greens')
plt.title('Has_Diabetes Rate by Age Group')
plt.ylabel('Average (Proportion)')
plt.show()
```

This bar chart reveals a strong positive correlation between age and the proportion of people with diabetes. The rate of diabetes rises with each successive age group, reaching its peak in the 75+ category.
